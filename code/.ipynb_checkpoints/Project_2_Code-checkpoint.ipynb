# Data Cleaning

### Dealing with Nulls

First, let's handle the missing values. We'll get all columns with missing values and take a look at their situtaions.

After storing all columns with nulls into a new DataFrame called 'null_cols', we discovered 26 such columns.

```python
# Creating a dataframe for columns containing at least one null
mask = df_train.isnull().sum() > 0
null_cols = df_train.loc[:, mask]
```

There are a few rows with close to or more than half of the values missing.

Let's start with Alley, Fireplace Qu, Fence, Misc Feature and Pool QC, all of which have more than 1000 missing values.

After doing some exploration in these data, we realized that the nulls in these cols actually had meanings. So we shouldn't delete these missing values and in turn assign them a different value to represent their true meaning.

```python
df_train.loc[:,df_train.isnull().sum() > 1000].columns
```

```python
# Based on the data description, null in Alley means 'no access'.
df_train['Alley'].fillna('NoAccess', inplace=True)
df_test['Alley'].fillna('NoAccess', inplace=True)

# Replace nulls with 'Ne', which meanings 'Not exist'
df_train['Fireplace Qu'].fillna('Ne', inplace=True)
df_test['Fireplace Qu'].fillna('Ne', inplace=True)

# Next, we take a look at fence.
# Null in fence means no fence, so we replace nulls by 'NoFence'.
df_train['Fence'].fillna('NoFence', inplace=True)
df_test['Fence'].fillna('NoFence', inplace=True)

# Next, misc feature.
# Null in misc feature means 'None', so we replace nulls with 'None'.
df_train['Misc Feature'].fillna('None', inplace=True)
df_test['Misc Feature'].fillna('None', inplace=True)

# Lastly, we move to Pool QC.
# Null in Pool QC means no pool, so we replace nulls with 'Np'.
df_train['Pool QC'].fillna('Np', inplace=True)
df_test['Pool QC'].fillna('Np', inplace=True)
```

Next, let's deal with the columns less than 10 missing values.
There are 6 total:
- Bsmt Unf SF
- BsmtFin SF 1
- BsmtFin SF 2
- Garage Area
- Bsmt Full Bath
- Bsmt Half Bath

For these, null simply are just missing values. Since quantity is small, let's just drop the rows. Do this only for df-train.

```python
df_train.dropna(subset=['Bsmt Unf SF', 
                        'BsmtFin SF 1', 
                        'BsmtFin SF 2',
                        'Garage Area',
                        'Bsmt Full Bath',
                        'Bsmt Half Bath'], inplace=True)
```

```python
# Opps, missed 2.
null_cols.dropna(subset=['Total Bsmt SF', 'Garage Cars'], inplace=True)
```

For Mas Vnr Type and Mas Vnr Area, since these nulls are actully due to entry problems and we have no good way to retrive those values, I think we can afford to lose these rows (22 at maximum).

```python
df_train.dropna(subset=['Mas Vnr Area', 'Mas Vnr Type'], inplace=True)
```

For Bsmt Qual, Bsmt Cond, BsmtFin Type 1, BsmtFin Type 2, Bsmt Exposure.

Null has meaning here. So we replace null in both training and test data with 'Nb'-No Basement.

```python
df_train['Bsmt Qual'].fillna('Nb', inplace=True)
df_test['Bsmt Qual'].fillna('Nb', inplace=True)

df_train['Bsmt Cond'].fillna('Nb', inplace=True)
df_test['Bsmt Cond'].fillna('Nb', inplace=True)

df_train['BsmtFin Type 1'].fillna('Nb', inplace=True)
df_test['BsmtFin Type 1'].fillna('Nb', inplace=True)

df_train['BsmtFin Type 2'].fillna('Nb', inplace=True)
df_test['BsmtFin Type 2'].fillna('Nb', inplace=True)

df_train['Bsmt Exposure'].fillna('Nb', inplace=True)
df_test['Bsmt Exposure'].fillna('Nb', inplace=True)
```

Same thing with garage.
Do the same thing as basements.
Fill nulls with 'Ng' as in 'No Garage'

```python
df_train['Garage Type'].fillna('Ng', inplace=True)
df_test['Garage Type'].fillna('Ng', inplace=True)

df_train['Garage Yr Blt'].fillna('Ng', inplace=True)
df_test['Garage Yr Blt'].fillna('Ng', inplace=True)

df_train['Garage Finish'].fillna('Ng', inplace=True)
df_test['Garage Finish'].fillna('Ng', inplace=True)

df_train['Garage Qual'].fillna('Ng', inplace=True)
df_test['Garage Qual'].fillna('Ng', inplace=True)

df_train['Garage Cond'].fillna('Ng', inplace=True)
df_test['Garage Cond'].fillna('Ng', inplace=True)
```

Lastly, for Lot Frontage, fill all nulls with 0. We are doing this because 330 rows of data is a lot to lose in a dataset with only 2000+ entries.

This is a feature we don't see as useful to include, so just imputing with 0 for some numeric placeholder since we cannot afford to lose 330 rows.

```python
df_train['Lot Frontage'].fillna(0, inplace=True)
df_test['Lot Frontage'].fillna(0, inplace=True)
```

Reset the index.

Check to see that all nulls are cleaned.

Moving to the next step.

```python
df_train.reset_index(drop=True, inplace=True)
df_train.isnull().sum().sort_values()
```

### Step 2: Balance for Categorical Variables

Even though we have a few graphics with messed up x labels, in general we are seeing what we want to see. There is a good amount of differences among the categories for each of our variables explored. This means that they might be very good predictors.

There is only one more thing I want to look at before going with this set of variables as our dummy candidates. 

We want the number of our predictors to be balanced. For example, if for a certain variable, the proportion/number of houses in different categories is tilted, we might not want to use that variable as a estimator.

```python
dummy_eda_list = ['Street', 
                  'Utilities',
                  'Neighborhood',
                  'House Style',
                  'Exter Qual',
                  'Foundation',
                  'Bsmt Exposure',
                  'Heating QC',
                  'Central Air',
                  'Kitchen Qual',
                  'Functional',
                  'Garage Qual',
                  'Pool QC',
                  'Sale Type']

rows = int(np.ceil(len(dummy_eda_list)/2))

fig, ax = plt.subplots(nrows=rows, 
                       ncols=2, 
                       figsize=(10, 30), 
                       sharex=False)

ax = ax.ravel()

plt.subplots_adjust(left = 0.125, 
                    bottom = 0.1,
                    top = 0.9,
                    wspace = 0.5, 
                    hspace = 0.5)  

for i in range(len(dummy_eda_list)):
    sns.barplot(data=df_train,
                x=dummy_eda_list[i],
                y='SalePrice',
                ci=None,
                estimator=len,
                ax=ax[i])
    ax[i].set_title(dummy_eda_list[i], fontsize=20)
    ax[i].set_ylabel('Count', fontsize=15)
    ax[i].set_xlabel(dummy_eda_list[i], fontsize=15)
```

We can quickly identify that the following variables do not have very good balance among the catgeories:

- Street
- Utilities
- Functional
- Pool QC

AS a result, we will drop these 4 and move on with the rest of the estimators. Below is our final list of dummy variables that we will run regression against.

```python
removing_list = ['Street', 'Utilities', 'Functional', 'Pool QC']

for i in removing_list:
    dummy_eda_list.remove(i)
```

# Processing

### Start with All Numerical Columns

We are going to start with the features below as they have shown the great correlation with the target. So let's set up the feature list based on these variables.
- Overall qual
- Gr Liv Area
- Garage Areas
- Total Bsmt SF
- 1st For SF
- Year Built
- Year Removed, added
- Full Bath
- Mas Vnr Area
- TotRms AbvGrd
- Fireplaces
- BsmtFin SF 1

First, let's do a log transformation on some of these variales. The following are great variable to transform. The following three variables are particular suitable for the transformation because:

1. They don't contain many zeros
2. The histogram after transformation shows distribution that resembles the normal distribution.

```python
# Above Ground Living Area Sqft
df_train['Gr Liv Area'] = (df_train['Gr Liv Area']).map(np.log)
df_test['Gr Liv Area'] = (df_test['Gr Liv Area']).map(np.log)
```

```python
plt.figure(figsize=(10,8))
sns.distplot(np.exp(df_train['Gr Liv Area']))
plt.xlabel('Above Ground Living Area', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.title('Histogram for Above Ground Living Area', fontsize=20, pad=20.0);
```

```python
# First Floor Square Feet
df_train['1st Flr SF'] = df_train['1st Flr SF'].map(np.log)
df_test['1st Flr SF'] = df_test['1st Flr SF'].map(np.log)
```

```python
# Total rooms above ground
df_train['TotRms AbvGrd'] = df_train['TotRms AbvGrd'].map(np.log)
df_test['TotRms AbvGrd'] = df_test['TotRms AbvGrd'].map(np.log)
```

```python
# Get all features with correlation above 0.4 against the target
mask = (num_cols.corr()['SalePrice'] > 0.4) & (num_cols.corr()['SalePrice'] < 1)
features = list(num_cols.corr()['SalePrice'][mask].index)

# Get our X
X = df_train[features]

# Get our y (target)
y = df_train['SalePrice']
```

### Explore Some Categorical Variables

```python
dummy_eda_list
```

### Create Dummies for Our Categorical Varibales

```python
# Getting dummy classes for a particular column
train_dummies = pd.get_dummies(df_train[dummy_eda_list])
test_dummies = pd.get_dummies(df_test[dummy_eda_list])

# Check to see which the discrepancy between training and testing data on those categories.
train_excl = set(train_dummies.columns) - set(test_dummies.columns)
test_excl = set(test_dummies.columns) - set(train_dummies.columns)

# Add all missing dummy cols to test dataset
for dummy_class in train_excl:
    test_dummies[dummy_class] = 0

# Add all missing dummy cols to train dataset
for dummy_class in test_excl:
    train_dummies[dummy_class] = 0

# Set the order of the dummies for both datasets
test_dummies = test_dummies[train_dummies.columns]
```

### Add Dummy Variables Back to `df_train`

```python
# Concat our dummy cols to df_train
df_train = pd.concat(objs=[df_train, train_dummies],
          axis=1)
```

```python
# Rearrange the columns to put 'SalePrice' at the last column.
col_arg = list(df_train.columns)
```

```python
col_arg.remove('SalePrice')
col_arg.append('SalePrice')
col_arg
df_train = df_train[col_arg]
```

### Add Dummy Variables Back to `df_test`

```python
# There is no need to reaggrange because 'SalePrice' is missing from df_test
df_test = pd.concat(objs=[df_test, test_dummies], axis=1)
```

### Construct Our New Feature List with Dummies

```python
dummy_cols = list(train_dummies.columns)
```

```python
features = features + dummy_cols
```

```python
X = df_train[features]
```

```python
y.shape
```

### Train-Test-Split

```python
# Now, the train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
# Take the log of the y.
y_train_pt = y_train.map(np.log)
y_test_pt = y_test.map(np.log)
```

### Scaling

```python
ss = StandardScaler()
ss.fit(X_train)
X_train_sc = ss.transform(X_train)
X_test_sc = ss.transform(X_test)
```

<a name='modeling'/>

# Modeling

### Baseline Model

```python
# Instantiate: creates a skeleton of a model that will always predict the mean of the target
base_mean = DummyRegressor(strategy='mean')

# Fit the (terrible) model
base_mean = base_mean.fit(X_train, y_train_pt)

# Get predictions!
y_hat_base_train = np.exp(base_mean.predict(X_train))
y_hat_base_test = np.exp(base_mean.predict(X_test))

# Evaluate: let's use MSE
print('The MSE for our baseline train data is {0:,.2f}'.format(mean_squared_error(y_train, y_hat_base_train)))
print('The MSE for our baseline test data is {0:,.2f}'.format(mean_squared_error(y_test, y_hat_base_test)))
```

### Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train, y_train_pt)
```

```python
lr.score(X_train, y_train_pt)
```

```python
# Use the model to predict y_hat for both train and test data
y_hat_lr_train = np.exp(lr.predict(X_train))
y_hat_lr_test = np.exp(lr.predict(X_test))

print('The train MSE for our model is: {0:,.2f}'.format(mean_squared_error(y_train, y_hat_lr_train)))
print('The test MSE for our model is: {0:,.2f}'.format(mean_squared_error(y_test, y_hat_lr_test)))
```

```python
cross_val_score(lr, X_train, y_train_pt, cv=5).mean()
```

```python
cross_val_score(lr, X_test, y_test_pt, cv=5).mean()
```

### Lasso Regression

```python
# Set up a list of Lasso alphas to check.
l_alphas = np.logspace(-3, 0, 100)

# Cross-validate over our list of Lasso alpha.
lasso = LassoCV(alphas=l_alphas,
                cv=5,
                max_iter=5000)

# Fit the model.
lasso.fit(X_train_sc, y_train_pt)
```

```python
lasso.alpha_
```

```python
lasso.score(X_train_sc, y_train_pt)
```

```python
X_test_sc.shape
```

```python
y_test_pt.shape
```

```python
lasso.score(X_test_sc, y_test_pt)
```

### Ridge Regression

```python
# Set up a list of alphas to check.
r_alphas = np.logspace(0, 5, 100)

ridge = RidgeCV(alphas=r_alphas, scoring='r2', cv=5)

ridge.fit(X_train_sc, y_train_pt)
```

```python
ridge.score(X_train_sc, y_train_pt)
```

```python
ridge.score(X_test_sc, y_test_pt)
```
